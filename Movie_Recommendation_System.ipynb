Importing The Libraries

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection And Pre-Proceesing

```python
df = pd.read_csv('/content/movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
df.info()
```

```python
for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
shared_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
print(shared_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(shared_features)
```

```python
print(feature_vectors)
```

```python
feature_vectors = vectorizer.fit_transform(shared_features)
```

Cosine Similarity

```python
similar = cosine_similarity(feature_vectors)
```

```python
print(similar)
```

```python
print(similar.shape)
```

```python
movie_name = input(' Enter your favourite movie name : ')
```

```python
list_of_all_titles = df['title'].tolist()
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
if not find_close_match:
    print("No close matches found for the movie name.")
else:
    close_match = find_close_match[0]
    print(close_match)
```

```python
index_of_the_movie = df[df.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similar[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
